# **Data Cleaning**

```python
df = df.drop("PassengerId", axis=1)
```

**Replace Age nulls by median because there are outliers in this column**

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
print(df["Age"].min())
```

```python
df.head()
```

**Drop Cabin column because it has 77.1% nulls, and we cannot replace or recover these nulls by any way**

```python
df = df.drop("Cabin",axis=1)
df
```
